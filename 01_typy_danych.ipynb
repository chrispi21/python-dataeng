```python
s0 = "\u0065"
s1 = "\u0119"  # "ę" - reprezentacja Unicode
s2 = "ę"  # "ę"
s3 = "e\u0328"  # "e" + ogonek
s4 = "\u0065\u0328"  # "e" w Unicode + ogonek

print(s0)
print(s1)
print(s2)
print(s3)

print(s1 == s2)
print(s1 == s3)
print(s1 == s4)
print(s3 == s4)


print(len(s0))
print(len(s1))
print(len(s2))
print(len(s3))
print(len(s4))
```

Dodatkowe operacje

Rozpakowywanie (ang. `unpacking`), czyli przypisanie zawartości krotki do kolejnych zmiennych:

```python
x, y, z = moja_krotka_1
print(x, y, z)
```

Jeśli nie potrzebujemy, wszystkich wartości:

```python
# *_ używamy, gdy nie potrzebujemy kolejnych elementów krotki
x, *_ = moja_krotka_1
print(x)
```

```python
# _ używamy, gdy nie potrzebujemy dowolnego elementu
x, _, y = moja_krotka_1
print(x, y)
```

```python
# *nazwa_listy używamy, gdy chcemy przypisać pozostałe elementy do listy
x, *inne = moja_krotka_1
print(x, inne)
```

Zliczanie wystąpień wybranego elementu:

Funkcja `zip` umożliwia łączenie krotek:

```python
tuple(zip(moja_krotka_1, moja_krotka_2))
```

Kolejne odpowiednie elementu krotek łączone są w pary. Elementy o indeksie 0 z `moja_krotka_1` i `moja_krotka_2` trafiają do 0 elementu nowej krotki itd.

Następnie możemy w prosty sposób rozpakować kolejne elementy naszej zagnieżdżonej krotki:

```python
macierz_krotek = tuple(zip(moja_krotka_1, moja_krotka_2))

for element_1, element_2 in macierz_krotek:
  print(element_1, element_2)
```

Na bazie słownika `uzytkownicy` utwórz słownik, którego kluczem będzie wiek użytkownika a wartościami lista nazw użytkowników.

```python
uzytkownicy = {
  "id_1": {"nazwa": "Feliks", "wiek": 30},
  "id_2": {"nazwa": "Obeliks", "wiek": 30},
  "id_3": {"nazwa": "Zosia", "wiek": 11},
  "id_4": {"nazwa": "Maja", "wiek": 11},
}
```

```python
wynik = {}
for u in uzytkownicy.values():
  lista_imion = wynik.get(u["wiek"], [])
  lista_imion.append(u["nazwa"])
  wynik[u["wiek"]] = lista_imion
print(wynik)
```

Dla chętnych

Która wersja rozwiązania powyższego ćwiczenia jest bardziej wydajna pod kątem utylizacji pamięci i czasu wykonania? Dla wyjaśnienia prześledź wyniki poniższych eksperymentów. Pomogą nam trzy funkcje, które służą do modyfikowania listy:

```python
def _append(n):
    wynik = []
    for i in range(n):
        wynik.append(i)
    return wynik


def _extend(n):
    wynik = []
    for i in range(n):
        wynik.extend([i])
    return wynik


def _plus(n):
    wynik = []
    for i in range(n):
        wynik = wynik + [i]
    return wynik
```

Eksperyment z czasem wykonania:

```python
import time

N = 10_000  # spróbuj też 50k, 100k

for func in [_append, _extend, _plus]:
    start = time.time()
    func(N)
    koniec = time.time()
    print(f"{func.__name__}: {koniec - start:.4f} s")
```

Eksperyment z zużyciem pamięci:

```python
import tracemalloc

N = 10_000  # spróbuj też 50k, 100k

for func in [_append, _extend, _plus]:
  tracemalloc.start()
  func(N)
  current, peak = tracemalloc.get_traced_memory()
  print(f"{func.__name__}; Current: {current / 10**6:.2f} MB; Peak: {peak / 10**6:.2f} MB")
  tracemalloc.stop()
```
